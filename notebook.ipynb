```python
stock = {
"pains au chocolat": 500,
"croissants": 400,
"baguettes": 500,
"beignets": 200
}
```

```python
def afficher_inventaire():
    print("Inventaire actuel :")
    for produit, quantite in stock.items():
        print(f"- {produit} : {quantite}")
```

```python
afficher_inventaire()
```
