```python
stock = {
"pains au chocolat": 500,
"croissants": 400,
"baguettes": 500,
"beignets": 200
}
```

```python
def afficher_inventaire():
    print("Inventaire actuel :")
    for produit, quantite in stock.items():
        print(f"- {produit} : {quantite}")
```

```python
afficher_inventaire()
```

```python
def fabriquer(produit, quantite):
    global stock
    
    # Si le produit n'existe pas encore, on le crée dans le stock
    if produit not in stock:
        stock[produit] = 0
    
    # On augmente la quantité
    stock[produit] += quantite
    
    print(f"Fabrication de {quantite} unités de '{produit}'.")
    print(f"Nouveau stock de {produit} : {stock[produit]}")
```

```python
fabriquer("croissants", 20)
```

```python
afficher_inventaire()
```
